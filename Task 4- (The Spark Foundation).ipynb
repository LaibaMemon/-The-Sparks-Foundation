
# Exploratory Data Analysis

```python
plt.subplots(figsize=(15,20))
sns.countplot('attacktype1_txt', data=df, palette='inferno', hue = df['targtype1_txt'])
plt.xticks(rotation=50)
plt.xlabel('Attack Type')
plt.title('Attacking Methods with their targets as hue')
plt.show()
```

```python
plt.subplots(figsize=(20,8))
sns.countplot('iyear', data=df, palette='crest', edgecolor=sns.color_palette('dark',7))
plt.xticks(rotation=90)
plt.xlabel('Years')
plt.title('Number Of Terrorist Attacks per year')
plt.show()
```

```python
print("Maximum Attacks mostly happened during the year of ",df['iyear'].value_counts().index[0])
```

```python
plt.subplots(figsize=(15,8))
sns.countplot('imonth',data=df,palette='YlOrBr',edgecolor=sns.color_palette('dark',7))
plt.xticks(rotation=90)
plt.xlabel('Months')
plt.title('Number Of Terrorist Attacks')
plt.show()
```

```python
print("Maximum Attacks mostly happened during the month of ",df['imonth'].value_counts().index[0])
```

```python
df_cp=df.copy()
df_cp['affected']=df['nkill']+df['nwound']
df_cp['affected'].value_counts()
```

```python
df1 = df_cp.sort_values(by='affected', ascending=False)[:20]
```

```python
heat =df1.pivot_table(index='country_txt', columns='iyear', values='affected')
heat.fillna(0,inplace=True)
```

```python
sns.heatmap(heat, cmap='PiYG')
plt.xlabel('Year')
plt.ylabel('Country')
plt.title('Heatmap for Attacked with respect to Country and Year')
```

```python
plt.figure(figsize=(15,6))
sns.barplot(df['region_txt'].value_counts()[:10].index,df['region_txt'].value_counts()[:10].values, palette='crest')
plt.title('No. of Terrorist Activities by Region', fontsize=20)
plt.xlabel('Regions', fontsize=16)
plt.ylabel('Count', fontsize=16)
plt.show()
```

```python
plt.subplots(figsize=(15,6))
sns.barplot(df['country_txt'].value_counts().index[:20],df['country_txt'].value_counts()[:20].values,palette='magma')
plt.title('Countries with most acts of terrorism')
plt.xlabel('Countries')
plt.ylabel('Count')
plt.xticks(rotation= 90)
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.barplot(df['city'].value_counts()[1:11].index,df['city'].value_counts()[1:11].values, palette="Blues")
plt.title('Cities with most acts of terrorism', fontsize=20)
plt.xlabel('Cities',fontsize=16)
plt.ylabel('Count',fontsize=16)
plt.show()
```

```python
plt.subplots(figsize=(20,5))
sns.countplot('targtype1_txt', data=df, palette='Blues')
plt.xticks(rotation=90)
plt.xlabel('Target Type')
plt.title('Targets by Terrorists')
plt.show()
```

```python
plt.figure(figsize=(18,10))
sns.barplot(df['gname'].value_counts()[1:16].values, df['gname'].value_counts()[1:16].index, orient='h', palette='magma')
plt.title('No. of Terrorist Activities by Terrorist Group', fontsize=20)
plt.xlabel('Groups', fontsize=16)
plt.ylabel('Count', fontsize=16)
plt.show()
```
